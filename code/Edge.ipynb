```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# 画像の読み込みと元画像の表示
house = cv2.imread('./../fig_raw/dog.jpg',0)
plt.imshow(house,cmap='gray')
plt.title('元画像')
plt.show()

# 微分フィルタ（横方向）
kernel = 0.5*np.array([[0, 0, 0], [-1, 0, 1], [0, 0, 0]])
house_hori = cv2.filter2D(house, -1, kernel)
# house_hori = np.where(house>0,0,255)
plt.imshow(house_hori)
plt.title('横方向に微分')
plt.show()

# 微分フィルタ（縦方向）
kernel = 0.5*np.array([[0, -1, 0], [0, 0, 0], [0, 1, 0]])
house_hori = cv2.filter2D(house, -1, kernel)
# house_hori = np.where(house>0,0,255)
plt.imshow(house_hori)
plt.title('横方向に微分')
plt.show()
# 木目（横方向）が強調されている

# ソーベルフィルタ
gray_x = cv2.Sobel(house, cv2.CV_32F, 1, 0, ksize=3)
gray_y = cv2.Sobel(house, cv2.CV_32F, 0, 1, ksize=3)
house_sobel = np.sqrt(gray_x ** 2 + gray_y ** 2).astype('uint8')
# house_sobel = np.where(house_sobel>0,0,255)
plt.imshow(house_sobel,cmap='gray')
plt.title('ソーベルフィルタ')
plt.show()

# ラプラシアンフィルタ
house_laplacian = cv2.Laplacian(house,cv2.CV_32F, ksize=25)
plt.imshow(house_laplacian,cmap='gray')
plt.title('ラプラシアンフィルタ')
plt.show()
```
