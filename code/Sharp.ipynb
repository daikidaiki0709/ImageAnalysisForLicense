```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# 
house = cv2.imread('./../fig_raw/wood_2.jpg',0)
plt.imshow(house,cmap='gray')
plt.title('')
plt.show()

# 
k = 20
kernel = np.array([[-k/9,-k/9,-k/9],[-k/9,((8*k/9)+1),-k/9],[-k/9,-k/9,-k/9]])
house_sharp = cv2.filter2D(house, -1, kernel)
plt.imshow(house_sharp,cmap='gray')
plt.title('')
plt.show()
```
