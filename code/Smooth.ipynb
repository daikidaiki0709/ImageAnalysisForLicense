```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# 
house_bgr = cv2.imread('./../fig_raw/dog.jpg')
house_rgb = cv2.cvtColor(house_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(house_rgb)
plt.title('')
plt.show()

# 
house_mean = cv2.blur(house_rgb,(25,25))
plt.imshow(house_mean)
plt.title('')
plt.show()

# 
house_gauss = cv2.GaussianBlur(house_rgb,(25,25),15)
plt.imshow(house_gauss)
plt.title('')
plt.show()

# 
house_median = cv2.medianBlur(house_rgb,25)
plt.imshow(house_median)
plt.title('')
plt.show()

# 
house_bilate = cv2.bilateralFilter(house_rgb,25,9,9)
plt.imshow(house_bilate)
plt.title('')
plt.show()
```
